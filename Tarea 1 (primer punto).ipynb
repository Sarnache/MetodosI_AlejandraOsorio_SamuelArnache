```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

*2.0.2 Máximos*

```python
ruta="EstrellaEspectro.txt"
df = pd.read_csv(ruta, delimiter="\s+", header=None, names=['x', 'y'])
df
```

```python
x=np.array(df["x"])
y=np.array(df["y"])
```

```python
plt.plot(x,y)
plt.show()
```

```python
dx=(x[-1]-x[0])/len(x)
yp=np.zeros_like(x)
```

```python
for i in range(len(x)):
    if i==0:
        yp[i]=(y[i+1]-y[i])/dx
    elif i==(len(x)-1):
        yp[i]=(y[i]-y[i-1])/dx
    else:
        yp[i]=(y[i+1]-y[i-1])/(2*dx)

```

```python
maximos_condicion = (yp[:-1] > 0) & (yp[1:] < 0)

maximos_indices = np.where(maximos_condicion)[0] + 1
maximos_x = x[maximos_indices]
maximos_y = y[maximos_indices]

plt.plot(x, y, label='Función original')
plt.scatter(maximos_x, maximos_y, color='red', label='Máximos', zorder=5)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

*2.0.5 Espiral de Arquímedes*

```python
tetha=np.linspace(0,2*np.pi,50)
a=0
b=1
```

```python
def espiral(a,b,tetha):
    return a+b*tetha

r=espiral(a,b,tetha)

plt.plot(tetha, r)
plt.xlabel('θ (radianes)')
plt.ylabel('r')
plt.title('Espiral de Arquímedes con a=0, b=1')
plt.show()
```

```python
def cartesiana_x(r,tetha):
    return r*np.cos(tetha)

def cartesiana_y(r,tetha):
    return r*np.sin(tetha)

x=cartesiana_x(r,tetha)
y=cartesiana_y(r,tetha)

plt.plot(x,y)
plt.xlabel("Coordenada cartesiana en $x$")
plt.ylabel("Coordenada cartesiana en $y$")
plt.show()
```
